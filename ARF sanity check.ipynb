```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
import xgboost as xgb
import plotly.express as px
```

```python
iris_gen = pd.read_csv("iris_gen.csv", index_col = 0)
iris_gen1 = pd.read_csv("iris_gen1.csv", index_col = 0)
iris_gen2 = pd.read_csv("iris_gen2.csv", index_col = 0)
iris_gen3 = pd.read_csv("iris_gen3.csv", index_col = 0)
iris_real = pd.read_csv("iris_real.csv", index_col = 0)
iris_df = pd.read_csv("iris_df.csv", index_col = 0)
```

```python
iris_real['label'] = 1
iris_gen['label'] = 0
X = pd.concat((iris_real, iris_gen))
X = X.reset_index().drop('index', axis = 1)
```

```python
X.tail()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse = False)
iris_enc = enc.fit_transform(X[['Species']])
X[['0', '1', '2']] = iris_enc
X.drop(['Species'], axis = 1, inplace = True)
X.insert(X.shape[1] - 1, 'label', X.pop('label'))
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X.iloc[:, :-1], X.iloc[:, -1], test_size = 0.1)
```

```python
model = xgb.XGBClassifier( n_estimators=20, max_depth = 3)
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
np.sum(preds == y_test)/len(y_test)
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 50, max_depth = 5)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
